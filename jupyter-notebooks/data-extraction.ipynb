## Trying to extract from Uniswap V3 subgraphs (https://thegraph.com/explorer?search=uniswap+v3)

```python
# Set the API endpoint URL
endpoint_url = 'https://api.thegraph.com/subgraphs/name/Uniswap/uniswap-v3'

# Set the GraphQL query
query = '''
query {
  pools(first: 1, where: {id: "0xCBCdF9626bC03E24f779434178A73a0B4bad62eD"}) {
    id
    poolDayData(orderBy: date, orderDirection: desc, first: 7) {
      date
      tvlUSD
    }
  }
}
'''

# Set the request headers
headers = {
    'Content-Type': 'application/json'
}

# Send the GraphQL request
response = requests.post(endpoint_url, json={'query': query}, headers=headers)

# Handle the response
if response.status_code == 200:
    data = response.json()
    tvl_data = data['data']['pools'][0]['poolDayData']

    # Process and analyze the TVL data as needed
    for entry in tvl_data:
        date = entry['date']
        tvl_usd = entry['tvlUSD']
        print(f"Date: {date}, TVL in USD: {tvl_usd}")
else:
    print('Error:', response.status_code)
```
